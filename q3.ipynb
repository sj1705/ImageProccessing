```python
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.image as mat_img
import numpy as np
import cv2
```

```python
def redeyeremove(img):
    origImg = np.array(img)
    # plt.imshow(img1)
    img1 = origImg.copy()
    for i in range(img1.shape[0]):
        for j in range(img1.shape[1]):
            if img1[i][j][0]>=0 and img1[i][j][1]>=0 and img1[i][j][2]>=140 and img1[i][j][0]<=220 and img1[i][j][1]<=100 and img1[i][j][2]<=255:
                img1[i][j] = [0,0,0]
    output = np.hstack((origImg,img1))
    plt.imshow(cv2.cvtColor(output,cv2.COLOR_BGR2RGB))
```

```python
img1 = cv2.imread(r"PA1_images\\kidred.jpg",1)
redeyeremove(img1)
```

```python
img2 = cv2.imread(r"PA1_images\\Red-eye_effect.png",1)
redeyeremove(img2)
```

```python
img3 = cv2.imread(r"PA1_images\\redeye.jpg",1)
redeyeremove(img3)
```

```python
img4 = cv2.imread(r"PA1_images\\redeyebaby.jpg",1)
redeyeremove(img4)
```
